# Project Title - Covid 19 Data Analysis 

```python
#import library for the analysis 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv data set 
data = pd.read_csv("covid19_Confirmed_dataset.csv")
data.head()
```

```python
# To check the shape of the data 
data.shape
```

```python
# Drop the useless column from  our data set 
data_1 = data.drop(["Lat","Long"],axis = 1,inplace = True)
```

```python
data.head()
```

```python
# Aggregate the data set on the country /region
covid_19 = data.groupby("Country/Region").sum()
```

```python
covid_19.head()
```

```python
#visual representation of data 
covid_19.loc["China"].plot()
covid_19.loc["India"].plot()
covid_19.loc["Spain"].plot()
plt.legend()
```

```python
 # first three days spreads 
covid_19.loc["China"][:3].plot()  
covid_19.loc["India"][:3].plot()
covid_19.loc["Spain"][:3].plot()
plt.legend()
```

```python
#derivative of the curve 
covid_19.loc["China"].diff().plot()
```

```python
#maximum infection rate 
covid_19.loc["China"].diff().max()
```

```python
# creating the list of maximum infection rate in the countries and add to the data set
country = list(covid_19.index)
max_infection_rate = []
for c in country:
    max_infection_rate.append(covid_19.loc[c].diff().max()) 
covid_19["max_infection_rate"] = max_infection_rate
```

```python
covid_19.head()
```

```python
# creating the new data frame for the finaal computation 
corona_data = pd.DataFrame(covid_19["max_infection_rate"])
```
